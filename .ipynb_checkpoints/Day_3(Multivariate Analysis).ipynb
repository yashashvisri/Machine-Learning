```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
df= pd.read_csv('Titanic-Dataset.csv')
```

```python
df.sample(5)
```

```python
# Scatterplot using Numerical Numerical cols
sns.scatterplot(data=df, x="Survived", y="Pclass")
```

```python
# Barplot(Numerical-Categorical)
sns.barplot(x=df['Survived'],y=df['Age'])
```

```python
#Multivariate (Numerical-Categorical)
sns.barplot(x=df['Pclass'],y=df['Age'],hue=df['Sex'])
```

```python
# Box-Plot(numerical-categorical)
sns.boxplot(x=df['Sex'],y=df['Age'], hue=df['Survived'])
```

```python
sns.distplot(df[df["Survived"]==0]['Age'], hist=False)
sns.distplot(df[df["Survived"]==1]['Age'], hist=False)
```
